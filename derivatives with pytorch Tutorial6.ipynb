```python
import torch
```

```python
#find instantanious change of y  where the value of that instantaneous point is x
x=torch.tensor(2.0, requires_grad=True) #req grad =true, else the computing gradient throws error
y=9*x**4 +2*x**3 + 3*x**2 + 6*x +1
y.backward()
x.grad
```

```python
# partial derivative
x=torch.tensor(1.0, requires_grad=True)
z=torch.tensor(2.0, requires_grad=True)
y= x**2 + z**3
y.backward()
x.grad
```
