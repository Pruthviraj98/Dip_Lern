```python
import torch
from torch.nn import  Linear
```

```python
torch.manual_seed(1) #generate seed for random numbers for linear class
model=Linear(in_features=1, out_features=1) #create a linear model
print(model.bias, model.weight)
```

```python
x=torch.tensor([[2.0], [3.3]])
print(model(x))
```
